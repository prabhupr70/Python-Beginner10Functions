### Adding & Removing Column(s) using in-built custom function


```python
def categorize_athlete(row):
    if row["height_cm"] < 175 and row["weight_kg"] < 70:
        return "Lightweight"
    elif row["height_cm"] < 185 and row["weight_kg"] < 80:
        return "Middleweight"
    else:
        return "Heavyweight"
    
bios["Category"] = bios.apply(categorize_athlete, axis=1) #For "axis", 1 means row & 0 means Column
bios
```
