```python
#Start filtering the data
#df[df["Product"] == "Electronics"]    
df[df["Category"] == "Electronics"]
df[(df["Category"] == "Electronics") & (df["Country"] == "USA")]    # AND Operator
df[(df["Category"] == "Electronics") | (df["Country"] == "USA")]    # OR Operator
```

```python
# MATH operator Operator

#df[(df["Quantity"] >   5)]    
df[(df["Quantity"] >= 15)]    
#df[(df["Quantity"] <   2)]   
#df[(df["Quantity"] !=  2)]   
```

```python
# Advanced Filter (CASE Sensitive)

df[df["CustomerName"].str.startswith("A")]
df[df["CustomerName"].str.endswith("a")]

df[df["Country"].isin(["USA", "Canada"])]
df[~df["Country"].isin(["USA", "Canada"])] # Works like NOT IN
```

```python
# Updating a row

df.loc[df["CustomerName"] == "Anna Ivanova", "Product"] = "Tim"
df.loc[df["CustomerName"] == "Anna Ivanova"]
```
